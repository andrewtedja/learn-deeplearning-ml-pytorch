## Looping

```python
cars = ["Ford", "Volvo", "BMW"]

for x in cars:
  print(x)
```

```python
# prompt: Buat looping angka yng di dalamnya ada conditional

# Looping angka dengan conditional
for i in range(1, 21):  # Loop dari 1 hingga 20 (inklusif)
  if i % 2 == 0:        # Jika angka genap
    print(f"{i} adalah bilangan genap")
  else:                 # Jika angka ganjil
    print(f"{i} adalah bilangan ganjil")
```

## Matplotlib

[Matplotlib](http://matplotlib.org/) adalah paket grafik yang paling umum digunakan, lihat [dokumentasinya](http://matplotlib.org/api/pyplot_api.html) untuk detail lebih lanjut, dan [contohnya](http://matplotlib.org/gallery.html#statistics) untuk inspirasi ide.

### Line Plots

```python
import matplotlib.pyplot as plt

x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 5, 4, 6, 4, 2]
plt.plot(x, y1, label="line L")
plt.plot(x, y2, label="line H")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Example")
plt.legend()
plt.show()
```

### Bar Plots

```python
# Lihat indeks 4 dan 6, yang menunjukkan kasus tumpang tindih.
x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

# Colors: https://matplotlib.org/api/colors_api.html

plt.bar(x1, y1, label="Blue Bar", color='b')
plt.bar(x2, y2, label="Green Bar", color='g')
plt.plot()

plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("Bar Chart Example")
plt.legend()
plt.show()
```

### Histograms

```python
import numpy as np

#Gunakan numpy untuk menghasilkan sejumlah data acak dalam kurva lonceng di sekitar nilai 5.
n = 5 + np.random.randn(1000)

m = np.arange(1000)
plt.bar(m, n)
plt.title("Raw Data")
plt.show()

plt.hist(n, bins=20)
plt.title("Histogram")
plt.show()

plt.hist(n, cumulative=True, bins=20)
plt.title("Cumulative Histogram")
plt.show()
```

### Scatter Plots

```python
x1 = [2, 3, 4]
y1 = [5, 5, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 2, 3, 4]
y3 = [6, 8, 7, 8, 7]

# Markers: https://matplotlib.org/api/markers_api.html

plt.scatter(x1, y1)
plt.scatter(x2, y2, marker='v', color='r')
plt.scatter(x2, y3, marker='^', color='m')
plt.title('Scatter Plot Example')
plt.show()
```

### Stack Plots

```python
idxes = [ 1,  2,  3,  4,  5,  6,  7,  8,  9]
arr1  = [23, 40, 28, 43,  8, 44, 43, 18, 17]
arr2  = [17, 30, 22, 14, 17, 17, 29, 22, 30]
arr3  = [15, 31, 18, 22, 18, 19, 13, 32, 39]

#Membuat legenda untuk grafik yang ditumpuk
plt.plot([], [], color='r', label = 'D 1')
plt.plot([], [], color='g', label = 'D 2')
plt.plot([], [], color='b', label = 'D 3')

plt.stackplot(idxes, arr1, arr2, arr3, colors= ['r', 'g', 'b'])
plt.title('Stack Plot Example')
plt.legend()
plt.show()
```

### Pie Charts

```python
labels = 'S1', 'S2', 'S3'
sections = [56, 66, 24]
colors = ['c', 'g', 'y']

plt.pie(sections, labels=labels, colors=colors,
        startangle=90,
        explode = (0, 0.1, 0),
        autopct = '%1.2f%%')

plt.title('Pie Chart Example')
plt.show()
```

### 3D Plots

```python
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x, y, z = axes3d.get_test_data()

ax.plot_wireframe(x, y, z, rstride = 2, cstride = 2)
#ax.plot_surface(x, y, z, rstride = 2, cstride = 2)

plt.title("Wireframe Plot Example")
plt.tight_layout()
plt.show()
```
