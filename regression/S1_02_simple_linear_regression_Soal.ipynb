```python
# Import all dependencies
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Read the data
```

```python
data = pd.read_csv("Advertising.csv")
```

```python
# Look at the data
```

```python
data.head() #melihat 5 data pertama dalam tabel
```

```python
# Remove redundant columns
```

```python
data.drop(['Unnamed: 0'], axis=1) #data.drop = drop specified labels from rows or columns. Link: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html, https://www.w3resource.com/pandas/dataframe/dataframe-drop.php
```

```python
# Visualise the data (TV Ads and Sales)
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(
    data['TV'],
    data['sales'],
    c='black'
)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
# Generate Linear Approximation of this data
```

```python
X = data['TV'].values.reshape(-1, 1) #Reshape with (-1, 1) = We have provided column as 1 but rows as unknown, so we get result new shape as (n, 1)
y = data['sales'].values.reshape(-1, 1)

# LinearRegression fits a linear model with coefficients w = (w1, â€¦, wp) to minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation.
# Link: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
reg = LinearRegression()
reg.fit(X, y) #Fit linear model.

print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
# Visualise the Best Fit Line
```

```python
predictions = reg.predict(X) #Returns predicted values. The return value = a numpy array that contains all the predicted values for the input values X.

plt.figure(figsize=(16, 8))
plt.scatter(
    data['TV'],
    data['sales'],
    c='black'
)

plt.plot(
    data['TV'],
    predictions,
    c='blue',
    linewidth=2
)

plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```
