```python
from sklearn.linear_model import LinearRegression
from sklearn.processing import PolynomialFeatures

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("Housing_train_test.csv")
X = data["area"].values.reshape(-1, 1)
y = data["price"]

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=42, random_state=26)
```

```python
model = LinearRegression()
poly = PolynomialFeatures(2)

poly.fit_transform(X)
model.fit(X_train, y_train)
```

```python
validation_data = pd.read_csv('Housing_validation.csv')
X_val = validation_data["area"].values.reshape(-1, 1)
y_val_pred = model.predict(X_val)
answer = pd.DataFrame({'area': X_val.flatten(), 'price': y_val_pred.flatten()})
answer.to_csv('Housing_prediction.csv', index=False)
```
