```python
import base64

# Fungsi untuk meng-encode gambar
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")


# Path  gambar Anda
image_path = "path gambar yang hendak diolah" # Ganti dengan path gambar yang anda inginkan

# Hasil konversi gambar ke string Base64
base64_image = encode_image(image_path)


response = client.responses.create(
    model="gpt-4.1",
    input=[
        {
            "role": "user",
            "content": [
                { "type": "input_text", "text": "apa yang ada di gambar ini?" },
                {
                    "type": "input_image",
                    "image_url": f"data:image/jpeg;base64,{base64_image}",
                },
            ],
        }
    ],
)

print(response.output_text)
```

```python
response = client.responses.create(
    model="gpt-4.1-mini",
    input="Buat gambar kucing belang-belang abu-abu yang sedang memeluk berang-berang yang menggunakan syal oranye", # Ganti dengan prompt yang kamu inginkan
    tools=[{"type": "image_generation"}],
)

# simpan gambar yang dihasilkan
image_data = [
    output.result
    for output in response.output
    if output.type == "image_generation_call"
]
    
if image_data:
    image_base64 = image_data[0]
    with open("otter.png", "wb") as f:
        f.write(base64.b64decode(image_base64))
```

```python
# Olah lebih lanjut gambar yang dihasilkan
response_fwup = client.responses.create(
    model="gpt-4.1-mini",
    previous_response_id=response.id,
    input="Buat lebih realistis gambar kucing dan berang-berang tersebut",# Ganti dengan prompt yang kamu inginkan`
    tools=[{"type": "image_generation"}],
)

image_data_fwup = [
    output.result
    for output in response_fwup.output
    if output.type == "image_generation_call"
]

if image_data_fwup:
    image_base64 = image_data_fwup[0]
    with open("cat_and_otter_realistic.png", "wb") as f:
        f.write(base64.b64decode(image_base64))
```
